# 基于物品的智能推荐算法

# <font color = 'black' size = 5 face = '微软雅黑'>分析目标</font>

## <font color = 'black' size = 3 face = '微软雅黑'>此部分数据分析报告的目的是基于物品的智能推荐算法，为目标用户5推荐其没看过但是可能会喜欢的电影。</font>

```python
df_table2 = df.pivot_table(columns = "电影id",index = '用户id',values = '评分')
df_table2
```

```python
corr_matrix2 = df_table2.corr(method = 'pearson',min_periods=10)
corr_matrix2
```

选择用户id=5的影迷作为我们的目标用户targetnum

所以接下来先提取targetnum看过的电影

```python
targetmovie = df_table2.loc[5].dropna()
targetmovie
```

```python
targetname = targetmovie.index
targetscore = targetmovie.values
```

```python
targetname
```

```python
targetname.values
```

```python
targetscore
```

```python
targetcorr_matrix=corr_matrix2[targetname].drop(index = targetname)
targetcorr_matrix
```

预测该用户将会给他没看过的电影打几分

```python
predict_score = targetcorr_matrix*targetscore
predict_score = predict_score.sum(axis = 1)
predict_score
```

```python
predict_score = predict_score.sort_values(ascending = False)
predict_score
```

```python
osusume = predict_score.index[0:5]
osusume = osusume.values
osusumelist = list(osusume)
osusumelist
```
